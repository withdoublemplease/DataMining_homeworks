**Задание 10. Ответьте на вопросы:**
    * Какая доля студентов сдала экзамен по математике (passmark >= 50)?
    * Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?
    * Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?

```python
# Какая доля студентов сдала экзамен по математике (passmark >= 50)?
pass_math = (df["math score"] >= passmark).sum() / df.shape[0]
# Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?
yes = (df["test preparation"] == "completed").sum()
pass_math_with_course = ((df["math score"] >= passmark)&(df["test preparation"] == "completed")).sum() / yes
# Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?
women_without_course = ((df["test preparation"] == "none")&(df["gender"] == "female")).sum()
women_without_course_and_math = ((df["test preparation"] == "none")&((df["gender"] == "female")&(df["math score"] < passmark))).sum() / women_without_course
print(pass_math, pass_math_with_course, women_without_course_and_math, sep = "\n")
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
df['Math_PassStatus'] = df["math score"].apply(lambda x: 'F' if x < 50 else 'P')
print("Количество студентов, сдавших экзамен по математике:", (df["Math_PassStatus"]=="P").sum())
print("Количество студентов, не сдавших экзамен по математике:", (df["Math_PassStatus"]=="F").sum())
df['Reading_PassStatus'] = df["reading score"].apply(lambda x: 'F' if x < 50 else 'P')
print("Количество студентов, сдавших экзамен по чтению:", (df["Reading_PassStatus"]=="P").sum())
print("Количество студентов, не сдавших экзамен по чтению:", (df["Reading_PassStatus"]=="F").sum())
df['Writing_PassStatus'] = df["writing score"].apply(lambda x: 'F' if x < 50 else 'P')
print("Количество студентов, сдавших экзамен по письму:", (df["Writing_PassStatus"]=="P").sum())
print("Количество студентов, не сдавших экзамен по письму:", (df["Writing_PassStatus"]=="F").sum())
df.head(5)
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
  if average_mark >= 90:
      return 'A'
  if average_mark >= 80:
      return 'B'
  if average_mark >= 70:
      return 'C'
  if average_mark >= 60:
      return 'D'
  if average_mark >= 50:
      return 'E'
  return 'F'
```

```python
df['Grade'] = df[['math score', 'reading score', "writing score"]].mean(axis=1).apply(GetGrade)
for i in set(df["Grade"]):
    print(f'Количество студентов, получивших оценку {i}:', (df["Grade"] == i).sum())
df.head(5)
```
